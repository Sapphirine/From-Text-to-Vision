```python
##divide frames of video into images: 
import cv2  
vc=cv2.VideoCapture('test.avi') 
c=1  
if vc.isOpened():  
    rval , frame = vc.read()  
else:  
    rval = False  
timeF=1 #number of skipped frames 
while rval:  
    rval,frame=vc.read()  
    if(c%timeF==0): #store one frame over timeF frames  
        cv2.imwrite('image/'+str(c)+'.jpg',frame) 
    c=c+1  
    cv2.waitKey(1)
vc.release()  
```

```python
##process the images with yolo,resnet and k-means
##retrive the label of each image based on average label distribution of the clusters generated by text processing 
##(generating without test images)
```

```python
## input boxes with labels
import cv2
import os
inp=open("boxes_labels.txt")
cur=''
dic={}
for line in inp.readlines():
    try:
        w=line.index('.jpg');
        cur=int(line[:w])
        dic[cur]=[]
    except:
        my=line.split()
        for i in range(0,4):
            my[i]=int(my[i])
        dic[cur].append(my)
inp.close()
```

```python
##draw boxes, labels and store
def mark(path1,path2,name,my):
    imagePath=path1+name
    image = cv2.imread(path2+name)
    resized = cv2.resize(image,(608,608))
    icv=resized
    for x in my:
        cv2.rectangle(icv, (x[0], x[2]), (x[1], x[3]), (255,0,0), 3)
        cv2.putText(icv,x[4], (x[0],x[2]-5), cv2.FONT_HERSHEY_SIMPLEX,0.5,2)
    cv2.imwrite(imagePath,icv)
```

```python
import os
img_root="results/"
img_out="predict/"
files=os.listdir(img_root)
for i in range(1,len(files)):
    w=int(i/timeF)+1
    if (i%timeF==0):
        w=w-1
    w=w*timeF
    you=dic[w]
    mark(img_out,img_root,str(i)+".jpg",you)
```

```python
##construct videos using images
def make_video(images, outimg=None, fps=30, size=None,
               is_color=True, format="XVID",outvid=None):
    from cv2 import VideoWriter, VideoWriter_fourcc, imread, resize
    fourcc = VideoWriter_fourcc(*format)
    vid = None
    for img in images:
        if vid is None:
            if size is None:
                size=img.shape[1], img.shape[0]
            vid=VideoWriter(outvid, fourcc, float(fps), size, is_color)
        print(size)
        if size[0]!=img.shape[1] and size[1]!=img.shape[0]:
            img=resize(img, size)
        vid.write(img)
    vid.release()
    return vid
```

```python
##reconstructing result video
im=[]
for i in range(1,len(files)):
    frame=cv2.imread(img_out+str(i)+'.jpg')
    im.append(frame)
make_video(im,outvid="test_out.avi")
```
